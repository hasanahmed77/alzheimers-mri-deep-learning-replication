```python
import os
from PIL import Image
import matplotlib.pyplot as plt

data_dir = "../data/"
classes = os.listdir(data_dir)

# filter out any hidden files (like .DS_Store on Mac)
classes = [cls for cls in classes if not cls.startswith(".")]

for cls in classes:
    class_folder = os.path.join(data_dir, cls)

    # filter to only common image extensions
    images = [
        f
        for f in os.listdir(class_folder)
        if f.lower().endswith((".png", ".jpg", ".jpeg"))
    ]

    if len(images) == 0:
        print(f"No images found in {class_folder}")
        continue

    sample_img = images[0]
    img_path = os.path.join(class_folder, sample_img)

    img = Image.open(img_path)
    plt.imshow(img, cmap="gray")
    plt.title(cls)
    plt.axis("off")
    plt.show()
    break
```
